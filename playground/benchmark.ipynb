```python
import matplotlib.pyplot as plt

# Benchmark names
benchmarks = [
    "Environment lookup chain",
    "Lambda application",
    "Tail recursion (factorial)",
    "Arithmetic sum 1..500",
    "Python interop: math.sqrt loop"
]

# Times in seconds
interpreter_times = [0.001471, 0.464375,  1.778183, 17.523907, 1.662593]
vm_times = [None, 0.267339, 0.538032, 5.360367, 0.630777]

# Plot setup
x = range(len(benchmarks))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 6))

# Bars
ax.bar([i - width/2 for i in x], interpreter_times, width, label="Interpreter", color='skyblue')
ax.bar([i + width/2 for i in x], [t if t is not None else 0 for t in vm_times], width, label="VM (exec only)", color='salmon')

# Labels
ax.set_ylabel("Time (s)")
ax.set_title("Benchmark: Interpreter vs VM Execution Time")
ax.set_xticks(x)
ax.set_xticklabels(benchmarks, rotation=30, ha='right')
ax.legend()

# Annotate bars with values
for i, t in enumerate(interpreter_times):
    ax.text(i - width/2, t + 0.05, f"{t:.3f}", ha='center', va='bottom', fontsize=9)

for i, t in enumerate(vm_times):
    if t is not None:
        ax.text(i + width/2, t + 0.05, f"{t:.3f}", ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Benchmark names (skip the one without VM time)
benchmarks = [
    "Lambda application",
    "Tail recursion (factorial)",
    "Arithmetic sum 1..500",
    "Python interop: math.sqrt loop"
]

# Interpreter and VM times
interpreter_times = [0.435413, 1.672662, 17.876200, 1.936275]
vm_times = [0.264481, 0.527670, 6.284893, 0.706513]

# Compute speedups
speedups = [i / v for i, v in zip(interpreter_times, vm_times)]

# Plot
fig, ax = plt.subplots(figsize=(10, 6))
x = range(len(benchmarks))
bars = ax.bar(x, speedups, color='lightgreen')

# Labels and title
ax.set_ylabel("Speedup (Interpreter ÷ VM)")
ax.set_title("Benchmark Speedup of VM over Interpreter")
ax.set_xticks(x)
ax.set_xticklabels(benchmarks, rotation=30, ha='right')

# Annotate bars with speedup values
for bar, s in zip(bars, speedups):
    ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.05,
            f"{s:.2f}×", ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```
